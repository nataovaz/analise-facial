```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("image/pessoas_camera.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
```

```python
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_RGB2GRAY)
classificador = cv2.CascadeClassifier("classificadores/haarcascade_frontalface_default.xml")
```

```python
faces = classificador.detectMultiScale(imagem_gray, 1.3, 5)
```

```python
faces[0]
```

```python
imagem_desenhada = imagem.copy()
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(imagem_desenhada, (x,y), (x+w, y+h), (255,255,0), 2)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(imagem_desenhada)
```

```python
face_imagem = 0

for (x, y, w, h) in faces:
    face_imagem += 1
    imagem_roi = imagem[y:y+h, x:x+w]
    imagem_roi = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)
    cv2.imwrite("face_" + str(face_imagem) + ".jpg", imagem_roi)
```
